# QS2021 USA Ranking 分析

## 爬虫

```python
import pandas as pd
import requests
import json
import re
```

```python
uni_name = "uni"
location = "location"
International_student_ratio_5 = "ind_14"
International_faculty_ratio_5 = "ind_18"
Faculty_Student_Ratio_20 = "ind_36"
Citations_per_faculty_20 = "ind_73"
Academic_reputation_40 = "ind_76"
Employer_reputation_10 = "ind_77"
one, two, three, four, five, six, seven, eight = [],[],[],[],[],[],[],[]
headers = {"User-Agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 11_1_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/88.0.4324.96 Safari/537.36"}
```

```python
response = requests.get("https://www.topuniversities.com/sites/default/files/qs-rankings-data/en/3740566_indicators.txt?1623183713?v=1623230726293",headers=headers,timeout=5)
qs_usa_dataset = json.loads(response.text)["data"]
```

```python
for i in qs_usa_dataset:
    one.append(re.match('.*class=\"uni-link\">(.*)</a></div></div>', i[uni_name]).group(1))
    two.append(re.match('.*\"td-wrap-in\">(.*)</div></div>', i[International_student_ratio_5]).group(1))
    three.append(re.match('.*\"td-wrap-in\">(.*)</div></div>', i[International_faculty_ratio_5]).group(1))
    four.append(re.match('.*\"td-wrap-in\">(.*)</div></div>', i[Faculty_Student_Ratio_20]).group(1))
    five.append(re.match('.*\"td-wrap-in\">(.*)</div></div>', i[Citations_per_faculty_20]).group(1))
    six.append(re.match('.*\"td-wrap-in\">(.*)</div></div>', i[Academic_reputation_40]).group(1))
    seven.append(re.match('.*\"td-wrap-in\">(.*)</div></div>', i[Employer_reputation_10]).group(1))
    eight.append(i[location])
```

```python
df = pd.DataFrame({
    "学校":one,
    "国家":eight,
    "国际学生":two,
    "国际教师":three,
    "师生比":four,
    "引用":five,
    "学术声誉":six,
    "雇主声誉":seven}
)
```

```python
df.index +=1
```

```python
df["国际学生"] = pd.to_numeric( df["国际学生"])
df["国际教师"] = pd.to_numeric( df["国际教师"])
df["师生比"] = pd.to_numeric( df["师生比"])
df["引用"] = pd.to_numeric( df["引用"])
df["学术声誉"] = pd.to_numeric( df["学术声誉"])
df["雇主声誉"] = pd.to_numeric( df["雇主声誉"])
```

## 参数调整

### 雇主声誉50% ，学术声誉50%

```python
df2 = df.copy(deep=True)
```

```python
df2["Scores"] = df2["雇主声誉"]*0.5 + df2["学术声誉"]*0.5
df2 = df2.sort_values("Scores",ascending=False)
df2 = df2.reset_index()
df2.index +=1
df2.rename(columns={"index":"原排名"},inplace=True) 
df2[df2["国家"] == "China (Mainland)"]
```

```python
df2[(df2["学校"] == "University of California, Los Angeles (UCLA)")| 
   (df2["学校"] == "New York University (NYU)")| 
   (df2["学校"] == "California Institute of Technology (Caltech)")]
```
